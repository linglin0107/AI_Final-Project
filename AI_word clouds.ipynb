Goal: we would like to see the affect of Covid pandemic from NLP aspect.
We would mainly use wordcloud to show the difference:
(1) comparation of wordclouds for yelp reviews before Covid and after Covid pandemic.
(2) comparation of wordclouds of covid banner for temporary closed and not closed restaurants.

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import pandas as pd
import os
import matplotlib.pyplot as plt
% matplotlib inline
```

#Compare the wordclouds of yelp reviews before and after covid pandemic

```python
yelpReview2020 = pd.read_csv('YelpReview_2020.csv').astype({'date': 'datetime64[ns]'}) 
yelpReview2020.head()
#yelpCovid = pd.read_csv('YelpCovid.csv')
```

##wordClouds before Covid

```python
# Create and generate a word cloud image:
wordcloud = WordCloud().generate(" ".join(yelpReview2020[yelpReview2020['date'] < '2020-01-01']['text']))

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
##wordClouds after Covid
```

```python
# Create and generate a word cloud image:
wordcloud = WordCloud().generate(" ".join(yelpReview2020[yelpReview2020['date'] >= '2020-01-01']['text']))

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

#compare the wordclouds of yelp banner between temporary closed and not temporary closed

```python
yelpState = pd.read_csv('yelpState.csv')  #.astype({'date': 'datetime64[ns]'}) 
yelpState.head()
```

#wordclouds of yelp Banner where restaurants are not temporary closed

```python
wordcloud = WordCloud().generate(" ".join(   \
            yelpState[yelpState['CovidBanner'] != 'FALSE'][yelpState['TemporaryClosedUntil'] == 'FALSE']['CovidBanner'].dropna()))
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

#wordclouds of yelp Banner where restaurants are temporary closed

```python
wordcloud = WordCloud().generate(" ".join(   \
            yelpState[yelpState['CovidBanner'] != 'FALSE'][yelpState['TemporaryClosedUntil'] != 'FALSE']['CovidBanner'].dropna()))
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
